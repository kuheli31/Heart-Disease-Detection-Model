#### 2. What types of data are present, and how should we handle them?
Classify features as:

#### 3. Are there missing values in the dataset, and how should we deal with them?

- Removing rows/columns with too many missing values.

#### 4. Where are the outliers, and why do they matter?
Outliers can:
- Skew statistical metrics (mean, standard deviation).
- Mislead machine learning models.

# Age vs. Maximum Heart Rate for Heart Disease

```python
# Maximum Heart Rate Count Achieved (thalach)
df["thalach"].value_counts()
```

```python
#Creating figure for visualization
plt.figure(figsize=(10,6))

#Scatter with positive examples
plt.scatter(df.age[df.target == 1],
            df.thalach[df.target == 1],
            color="salmon")

#Scatter with negative examples
plt.scatter(df.age[df.target == 0],
            df.thalach[df.target == 0],
            color="blue")

#Adding info of the datas
plt.title("Heart Disease in function of Age and Max Heart Rate")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate")
plt.legend(["Heart Disease" , "No Heart Disease"])
plt.show()
```

# Heart Disease Frequency per Chest Pain Type

Chest pain (cp) type (0–3):

* 0 - Typical angina (chest pain due to decreased blood supply to the heart)
* 1 - Atypical angina (chest pain not related to the heart)
* 2 - Non-anginal pain (typically esophageal spasms)
* 3 - Asymptomatic: chest pain not showing signs of disease

```python
#Cross-tabulation between chest pain type and heart disease target
pd.crosstab(df.cp , df.target)
```

```python
#Making the crosstab more visual
pd.crosstab(df.cp , df.target).plot(kind="bar",
                                    figsize=(10,6),
                                    color=["lightblue" , "red"])

#Adding info of the datas
plt.title("Heart Disease Frequency per chest pain type")
plt.xlabel("Chest Pain Type")
plt.ylabel("Amount")
plt.legend(["No Heart Disease" , "Heart Disease"])
plt.xticks(ticks=[0, 1, 2, 3], 
           labels=[
               "0 = Typical Angina",
               "1 = Atypical Angina",
               "2 = Non-anginal Pain",
               "3 = Asymptomatic Pain"
           ],
           rotation=0)
plt.show()
```
